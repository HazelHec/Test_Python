<h1>Tarea Grafica 'Beggs and Brill'</h1>

### Datos requeridos para el calculo y librerias 
 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#NOTE Variables
Tpr = list(float(round(x,2)) for x in np.arange(1.2,2.5,0.1))
Ppr = list(float(round(x,2)) for x in np.arange(1,12.5,0.5))
```

<span style='color: #f5b041; font-size: 25px;'>Methods</span>

```python
def z_method(Tpr : float, Ppr : float): 
    A = 1.39 * pow((Tpr - 0.92),0.5) - 0.36 * Tpr - 0.10
    E = 9 * (Tpr - 1)
    B = (0.62 - 0.23 * Tpr) * Ppr + ((0.066 / (Tpr - 0.86)) - 0.037) * pow(Ppr,2) + ((0.32 * pow(Ppr,6))/pow(10,E))
    C = 0.132 - 0.32 *  np.log10(Tpr)
    F = 0.3106 - 0.49 * Tpr + 0.1824 * pow(Tpr,2)
    D = pow(10, F)
    
    return A + ((1-A)/np.exp(B)) + C * pow(Ppr, D)
    
```

<h3 style = "color: #ff5733;">DataFrame</h3>
<h5>Columnas : Tpr <br> Filas : Ppr </h5>

```python
data = { x : list(z_method(x,y) for y in Ppr) for x in Tpr}
dtf = pd.DataFrame( data = data, index=Ppr)
dtf
```

## Create Graphic

```python
# Create Graphic
graph, axes = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
for column in dtf.columns:
    axes.plot(dtf.index, dtf[column], '>-', label=f'Tpr : {column}')
axes.set_title('Beggs and Brill')
axes.set_xlabel('Ppr', labelpad=12.5)
axes.set_ylabel('Z', labelpad=12.5)
axes.legend()
graph.show()
```
